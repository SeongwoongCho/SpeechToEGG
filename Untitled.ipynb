```python
import librosa
import numpy as np
```

```python
[S,E],sr = librosa.load("./datasets/TrainData/CMU/bdl_arctic_a0080.wav",sr=16000,mono=False)
```

```python
def normalize(x):
    return -1+ 2 *(x-np.min(x))/(np.max(x)-np.min(x))

def get_power(x):
    return np.mean(x*x)
def get_power_list(a):
    n_frame = 200
    step = 30
    result = []
    for i in range(0,len(a)-n_frame,step):
        result.append(get_power(a[i:i+n_frame]))
    return result
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
S_p = get_power_list(normalize(S))
E_p = get_power_list(normalize(E))
```
